### C.) Read all product sales data and show it using a multiline plot. Label them

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the provided data
data = {
    'Brand': ['Apple', 'Samsung', 'Google', 'Sony', 'Microsoft', 'Dell', 'HP', 'Lenovo', 'Asus', 'Acer'] * 3,
    'Device': ['iPhone', 'Galaxy', 'Pixel', 'PlayStation', 'Surface', 'XPS', 'Spectre', 'ThinkPad', 'ROG', 'Predator'] * 3,
    'Price (USD)': [1099, 1199, 899, 499, 999, 999, 1349, 1429, 1499, 1299,
                    1999, 1699, 649, 1299, 499, 1999, 749, 999, 799, 699,
                    1099, 849, 229, 349, 899, 499, 1799, 1499, 1199, 599]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(12, 6))

for brand in df['Brand'].unique():
    brand_data = df[df['Brand'] == brand]
    plt.plot(brand_data['Device'], brand_data['Price (USD)'], marker='o', label=brand)

plt.xlabel('Device')
plt.ylabel('Price (USD)')
plt.title('Product Sales Data by Brand')
plt.xticks(rotation=45)
plt.legend(title='Brand')
plt.grid()
plt.tight_layout()
plt.show()
```

### D.) Read each brand sales data and show it using the bar chart. Label them.

```python
x = data['Brand']
y = data['Price (USD)']
plt.figure(figsize=(10, 6))
plt.bar(x, y)
plt.xlabel('Brand')
plt.ylabel('Total Sales (USD)')
plt.title('Total Sales by Brand')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

### E.) Read each device and show it using the histogram to see the most common Country of Origin. Label them.

```python
g = np.array([14, 3, 2, 4, 6, 1])
Origin = ["United States", "South Korea", "Japan", "China", "Taiwan", "Malaysia"]

# Create a histogram-like bar plot
plt.hist(Origin, weights=g, bins=len(Origin), edgecolor='black')
plt.xlabel('Country of Origin')
plt.ylabel('Count')
plt.title('Count of Devices by Country of Origin')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

###F.) Create a pie chart that shows each brand Country of Origin. Label them

```python
g = np.array([5, 1, 1, 2, 2, 1])
Origin = ["United States", "South Korea", "Japan", "China", "Taiwan", "Malaysia"]


plt.pie(g, labels=Origin, autopct='%1.1f%%')
plt.title('Country of Origin for Devices')
plt.legend(title="Country of Origin", labels=Origin, loc="upper left")
plt.show()
```
