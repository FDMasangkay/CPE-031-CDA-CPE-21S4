```python
physical_health_sum = data['PhysicalHealthDays'].sum()
mental_health_sum = data['MentalHealthDays'].sum()

labels = ['Physical Health Days', 'Mental Health Days']
sizes = [physical_health_sum, mental_health_sum]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Distribution of Physical and Mental Health Days')
plt.show()

'''
## 1.) Which state does heart attack oocur the most?

Based on the data, the state of Washington has the highest occurence of heart attack.
The bar graph greatly visualise the number of hadHeartAttack as the graph display
the output side to side to another Country's state
'''
```

```python
state_heart_attack = data.groupby('State')['HadHeartAttack'].count()

plt.figure(figsize=(15, 6))
plt.bar(state_heart_attack.index, state_heart_attack.values)
plt.xlabel("State")
plt.ylabel("Number of Heart Attacks")
plt.title("Number of Heart Attacks per State")
plt.xticks(rotation=90)
plt.show()

'''
## 2.) Does the respondents prioritize either Physical and Mental health aspects of their life?

- Either of it was prioritized, the pie chart shows that there's a balance on the occurence of it.
 I would say that respondents regardless of their gender did prioritize both Mental and Physical aspect in their life.
'''
```
