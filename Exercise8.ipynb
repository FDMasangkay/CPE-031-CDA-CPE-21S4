### Matplot Labels and TItle - You can use the xlabel() and ylabel()

```python
xpoints = np.array ([1, 2, 6, 8]) # plotting x and y points
ypoints = np.array ([3, 8, 1, 10])

plt.plot(xpoints, ypoints, color = 'yellow')# plot () is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") # labels
plt.ylabel("Speed in km/s")
plt.show()
```

### Matplot Grid - You can use grid () to add grid lines

```python
# select grid lines to display
xpoints = np.array ([1, 2, 6, 8]) # plotting x and y points
ypoints = np.array ([3, 8, 1, 10])

plt.plot(xpoints, ypoints)# plot () is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") # labels
plt.ylabel("Speed in km/s")
plt.grid(axis = 'y')
plt.show()
```

### Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot (x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Box Plot

```python
import matplotlib.pyplot as plt

data = [[1.5] * 10 + [2] * 10 + [3] * 10]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

### Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie (sizes, labels = labels)
plt.title("Pie Chart Example")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
champion = np.array([
    "Aatrox",
    "Alistar",
    "Amumu",
    "Blitzcrank",
    "Braum"
])

# Attack Speed Data
attack_speed = np.array([
    0.651,
    0.625,
    0.736,
    0.625,
    0.644
])

# Damage Data
damage = np.array([
    750,
    675,
    720,
    690,
    740
])

# Create subplots
plt.figure(figsize=(12, 5))

# Plot 1: Attack Speed
plt.subplot(1, 2, 1)
plt.plot(champion, attack_speed, label='Attack Speed', color='green', marker='x')
plt.title("Champion Attack Speed")
plt.xlabel("Champions")
plt.ylabel("Attack Speed")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend()

# Plot 2: Damage
plt.subplot(1, 2, 2)
plt.plot(champion, damage, label='Damage', color='red', marker='o')
plt.title("Champion Damage")
plt.xlabel("Champions")
plt.ylabel("Damage")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
